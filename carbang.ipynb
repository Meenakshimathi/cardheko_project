```python
import pandas as pd

# Normalize the 'new_car_detail' column
new_car = pd.json_normalize(df_combined['new_car_detail'])

# Display the first row of the new_car DataFrame
new_car.head(1)
```

```python
def row_find(row):
    row_info={item['key']:item['value'] for item in row ['top']}
    return row_info
view_unpack=df_combined['new_car_overview'].apply(lambda row:pd.Series(row_find(row)))
view_unpack.head(1)
```

```python
def p_features(row):
    features = {}

    # Flatten 'top' values
    for item in row['top']:
        features[item['value']] = True
    
    # Flatten 'data' values
    for data_item in row.get('data', []):
        for sub_item in data_item.get('list', []):
            features[sub_item['value']] = True
    
    return features

features_unpack= df_combined['new_car_feature'].apply(lambda row: pd.Series(p_features(row)))

features_unpack.head(1)
```

```python
import pandas as pd

def find_carspecs(row):
    spec = {}
    
    # Extract specifications from 'top' key
    for item in row.get('top', []):
        key = item.get('key')
        value = item.get('value')
        if key and value:  # Check if key and value are not None
            spec[key] = value
    
    # Extract specifications from nested 'data' key
    for data_item in row.get('data', []):
        for sub_item in data_item.get('list', []):
            key = sub_item.get('key')
            value = sub_item.get('value')
            if key and value:
                spec[key] = value

    return spec

# Apply the modified function and expand the dictionary into separate columns
spec_unpack = df_combined['new_car_specs'].apply(lambda row: pd.Series(find_carspecs(row)))

# Display the first row to check if the specs have been unpacked correctly
print(spec_unpack.head(1))
```
